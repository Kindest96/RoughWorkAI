```python
import csv
import numpy as np

data = []

with open('sonar.csv', 'r') as read_file:
    raw_data = csv.reader(read_file)
    for rows in raw_data:
        data.append(rows)
    data = np.array(data)

np.random.shuffle(data)

train_data = data[0:188, :60]
train_labels = data[0:188, -1]

test_data = data[188:208, :60]
test_labels = data[188:208, -1]

#Mine is 1 and Rock is 0
def encode_m_or_r(m_or_r):
    result = np.zeros((len(m_or_r)))
    for i, row in enumerate(m_or_r):
        if row == 'M':
            result[i] = 1.
    
    return result

x_train = train_data.astype('float32')
x_test = test_data.astype('float32')

y_train = encode_m_or_r(train_labels)
y_test = encode_m_or_r(test_labels)

print(x_train, y_train, x_test, y_test)
```

```python
from keras import models
from keras import layers

network = models.Sequential()
network.add(layers.Dense(1024, activation='relu', input_shape=(60,)))
network.add(layers.Dense(1024, activation='relu'))
#network.add(layers.Dense(256, activation='relu'))
#network.add(layers.Dense(256, activation='relu'))
network.add(layers.Dense(1, activation='sigmoid'))

network.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

history = network.fit(x_train, y_train, epochs=100, batch_size=512)

network.evaluate(x_test, y_test)
```

```python
import matplotlib.pyplot as plt

history_dict = history.history
loss_values = history_dict['loss']
#val_loss_values = history_dict['val_loss']

epochs = range(1, len(history_dict['accuracy']) + 1)
plt.plot(epochs, loss_values, 'b', label='Training Loss')
#plt.plot(epochs, val_loss_values, 'b', label='Validation Loss')
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
plt.clf()

accuracy_values = history_dict['accuracy']
#val_accuracy_values = history_dict['val_accuracy']

plt.plot(epochs, accuracy_values, 'b', label='Training Accuracy')
#plt.plot(epochs, val_accuracy_values, 'b', label='Validation Accuracy')
plt.title('Training Accuracy')
plt.xlabel('Accuracy')
plt.ylabel('Epochs')
plt.legend()

plt.show()
```
